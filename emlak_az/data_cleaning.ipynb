```python
# Assuming you have already loaded the DataFrame, and it's named df

# Split the phone numbers in the "phone_numbers" column
df['phone_numbers_2'] = df['phone_numbers'].str.split(', ')

# Explode the list of phone numbers into separate rows
df_exploded = df.explode('phone_numbers_2')

# Reset the index of the DataFrame
df_exploded.reset_index(drop=True, inplace=True)

```

```python
df_exploded.to_csv(r"data2.csv", sep=',', encoding='utf-8', header='true')
df_exploded.to_parquet(r"data2.parquet")
df_exploded.to_excel('data2.xlsx', engine='xlsxwriter')
```
